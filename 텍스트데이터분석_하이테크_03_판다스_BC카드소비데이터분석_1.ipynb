### 2. 서울시 거주/비거주 고객 비교 분석

### 2-1 서울시 거주/비거주(seoul, non_seoul)

```python
seoul = bc_df[bc_df['CSTMR_MEGA_CTY_NM'] == '서울특별시']
non_seoul = bc_df[bc_df['CSTMR_MEGA_CTY_NM'] != '서울특별시']
```

### 2-2 고객 수, 총 소비액, 카드 이용 건수

```python
print(f"서울시 거주 고객수 : {len(seoul):,} 명\n서울시 비거주 고객수 : {len(non_seoul):,} 명")
print(f"서울시 거주 총 소비액 : {seoul['AMT'].sum():,} 원\n서울시 비거주 총 소비액 : {non_seoul['AMT'].sum():,} 원")
print(f"서울시 거주 총 이용건수 : {seoul['CNT'].sum():,} 건\n서울시 비거주 총 이용건수 : {non_seoul['CNT'].sum():,} 건")
```

## 3. 편의점 소비 정보 분석

### 3-1 편의점 소비액, 강남구 소비액

```python
cstore_total_amt = bc_df[bc_df['TP_BUZ_NO'] == 4010]['AMT'].sum()
gangnam_total_amt = bc_df[bc_df['CTY_RGN_NM'] == '강남구']['AMT'].sum()
print(f"편의점 소비액 : {cstore_total_amt:,} 원")
print(f"강남구 소비액 : {gangnam_total_amt:,} 원")
```

### 3-2 강남구 편의점 소비액

```python
gangnam_cstore_df = bc_df[(bc_df['CTY_RGN_NM'] == '강남구') & (bc_df['TP_BUZ_NO'] == 4010)]
gangnam_cstore_amt = gangnam_cstore_df['AMT'].sum()
print(f"강남구 편의점 소비액 : {gangnam_cstore_amt:,} 원")
```

### 3-3 강남구 편의점 소비액 중 강남구 거주자와 비거주자의 소비액 비교

```python
resident_cstore_amt = gangnam_cstore_df[gangnam_cstore_df['CSTMR_CTY_RGN_NM'] == '강남구']['AMT'].sum()
non_resident_cstore_amt = gangnam_cstore_df[gangnam_cstore_df['CSTMR_CTY_RGN_NM'] != '강남구']['AMT'].sum()
print(f"강남구 편의점 소비액: {gangnam_cstore_amt:,.0f}원")
print(f"강남구 거주자 편의점 소비액: {resident_cstore_amt:,} 원")
print(f"강남구 비거주자 편의점 소비액: {non_resident_cstore_amt:,} 원")
```
